## Yearly

```python
df_brand_yearly_performance = pd.read_excel('brand_yearly_performance.xlsx')
df_brand_yearly_performance = df_brand_yearly_performance.drop(['Unnamed: 0'], axis=1)
```

```python
brands = list(set(df_brand_yearly_performance['brand'].to_list()))
year = list(set(df_brand_yearly_performance['year'].to_list()))
```

```python
# handling missing years
missing_years = []
num_reviews = []
missing_brands = []
for brand in brands:
    interim_df = df_brand_yearly_performance[df_brand_yearly_performance['brand'] == brand]
    missing_dates_len = len(year) - len(interim_df)
    if missing_dates_len != 0:
        for i in range(missing_dates_len):
            missing_brands.append(brand)
            missing_years.append(1999+i)
            num_reviews.append(0)

merged_list = list(zip(missing_brands, missing_years, num_reviews))
df_missing_values = pd.DataFrame(merged_list, columns=['brand', 'year', 'number_of_reviews'])
```

```python
complete_df_yearly_brand = pd.concat([df_brand_yearly_performance,df_missing_values], ignore_index=True)
```

```python
for brand in brands:
    inter_df = complete_df_yearly_brand[complete_df_yearly_brand['brand'] == brand].sort_values('year')
    number_of_reviews = inter_df['number_of_reviews']
    plt.plot(year, number_of_reviews, label=brand)

plt.xlabel('Year')
plt.ylabel('Number of reviews')
plt.title('Number of reviews of top 10 brands by year')
plt.legend(bbox_to_anchor=(1.5, 1), loc='upper left', borderaxespad=0)
plt.savefig('brand_year_performance.png')
plt.show()
```
